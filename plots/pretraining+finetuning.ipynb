```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
output_dir = "../output/pretraining/20221025-155605"

df = pd.read_csv(os.path.join(output_dir, "training.log"))
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].plot(df["epoch"], df["accuracy"], label="train")
ax[0].plot(df["epoch"], df["val_accuracy"], label="val")
ax[0].set_xlabel("Epochs")
ax[0].set_ylabel("Train/Val Accuracy")
ax[0].legend()
ax[0].set_title(f"Train/Val Accuracy against Epochs \nfor Pretraining (on CelebA)")

ax[1].plot(df["epoch"], df["loss"], label="train")
ax[1].plot(df["epoch"], df["val_loss"], label="val")
ax[1].set_xlabel("Epochs")
ax[1].set_ylabel("Train/Val Loss")
ax[1].legend()
ax[1].set_title(f"Train/Val Loss against Epochs \nfor Pretraining (on CelebA)")
```

```python
output_dir = "../output/finetuning/20221026-220936/"

for i in range(5):
    df = pd.read_csv(os.path.join(output_dir, f"fold_{i}", "training.log"))
    print(max(df["val_accuracy"]))
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))
    ax[0].plot(df["epoch"], df["accuracy"], label="train")
    ax[0].plot(df["epoch"], df["val_accuracy"], label="val")
    ax[0].set_xlabel("Epochs")
    ax[0].set_ylabel("Train/Val Accuracy")
    ax[0].legend()
    ax[0].set_title(f"Train/Val Accuracy against Epochs for fold {i}")

    ax[1].plot(df["epoch"], df["loss"], label="train")
    ax[1].plot(df["epoch"], df["val_loss"], label="val")
    ax[1].set_xlabel("Epochs")
    ax[1].set_ylabel("Train/Val Loss")
    ax[1].legend()
    ax[1].set_title(f"Train/Val Loss against Epochs for fold {i}")

    plt.show()
```
