```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
output_dir = "../output/supcon_pretraining/20221109-092421"


df = pd.read_csv(os.path.join(output_dir, "training.log"))
plt.figure(figsize=(5, 4))
plt.plot(df["epoch"], df["loss"])
plt.xlabel("Epochs")
plt.ylabel("SupCon Loss")
plt.title("SupCon Loss against Epochs for the pretraining stage in SCL")
```

```python
df = pd.read_csv(os.path.join(output_dir, "training.log"))
plt.figure(figsize=(5, 4))
plt.plot(df["epoch"], df["lr"])
plt.xlabel("Epochs")
plt.ylabel("Learning Rate")
plt.title("Learning Rate against Epochs for the pretraining stage in SCL")
```
